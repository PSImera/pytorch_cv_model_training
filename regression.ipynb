# Imports

```python
import numpy as np
import matplotlib.pyplot as plt

import torch
from torch import nn

from torch.optim.lr_scheduler import ReduceLROnPlateau as Sheduler
torch.manual_seed(1337)

from src.datasets import REGRDataset
from src.model import MyModel, FC2LNN
from src.utils import EarlyStopping

plt.style.use('dark_background')

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

# Initializing Dataset

```python
dataset_reg = REGRDataset('dataset')
print(f'Dataset initialized. len = {len(dataset_reg)}', end='\n\n')

rand_i = np.random.randint(0, len(dataset_reg))
img, coords = dataset_reg.view(rand_i)
print(f'img[{rand_i}]:')

print(f'Center Coordinates: {coords}')
plt.figure(figsize=(5, 5))
plt.scatter(coords[0], coords[1], color='red', marker='x')
plt.imshow(img, cmap='gray')
plt.show()
```

# Initializing Models

```python
# model = MyNN(in_channels=1, output=2, activator='relu', input_size=(64, 64))
model = FC2LNN(64*64, 2)
opt = torch.optim.Adam(model.parameters(), lr=0.001)
sheduler = Sheduler(opt, mode='min', factor=0.1, patience=4, threshold=0.001)
earlystoper = EarlyStopping(mode='min', patience=8, threshold=0.001, threshold_mode='rel')

regression = MyModel(
    dataset=dataset_reg,
    model=model,
    optimizer=opt,
    loss_function=nn.MSELoss(),
    sheduler=sheduler,
    earlystoper=earlystoper,
    split_list=[0.8, 0.10, 0.10],
    is_reshape=True,
    batch=64,
    save_path='models',
    name='reg',
    device=device,
    classification=False,
    )
```

# Train

```python
regression(60)
```

# 

```python
regression.load_model('models\\reg_ckpt_epoch_57.pt')
print(f'Regression model test')
test_acc, test_loss = regression.test()
```

# Inference

```python
rand_i = np.random.randint(0, len(dataset_reg))
pred = regression.inference(dataset_reg[rand_i][0])
img, true_coords = dataset_reg.view(rand_i)

print(f'True: {true_coords}')
print(f'Predict: {pred}')

plt.figure(figsize=(5, 5))
plt.scatter(true_coords[0], true_coords[1], color='green', marker="o")
plt.scatter(pred[0], pred[1], color='red', marker="x")
plt.imshow(img, cmap='gray')
plt.show()
```
